```python
#import libaries needed and add comments -
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# remove row that contains null values - Adanma Iheanacho
df.dropna()
```

```python
# outliers - Adanma Iheanacho

# Create a boxplot for each numeric column
for col in df.select_dtypes(include=["int", "float"]):
    sns.boxplot(x=df[col])
    plt.show()

# Calculate the z-score for each numeric column
for col in df.select_dtypes(include=["int", "float"]):
    z_scores = (df[col] - df[col].mean()) / df[col].std()

    # Identify outliers based on a threshold (e.g., z-score > 3)
    outliers = df[abs(z_scores) > 3]

    # Print the number of outliers for each column
    print(f"Number of outliers in {col}: {outliers.shape[0]}")
```

# Data Exploration
Let's initiate an exploratory analysis to delve into the intricacies of the population dataset and extract meaniful insights.

```python
#Computing Pairwise correlation - Khadijat
df.corr()
```

```python
#plotting the heatmap - Khadijat
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

```python
#plot a bar chart showing 2022 population by continent - Khadijat

df.groupby(['Continent']).sum()['2022 Population'].sort_values(ascending=False).plot(kind='bar', color='maroon', title= '2022 Population Distribution by Continents', y='Population in Billions')
```

```python
  plt.figure(figsize=(8, 6))
  sns.violinplot(x='Continent', y='2022 Population', data=df)
  plt.title('Population Distribution by Continent')
  plt.show()
```

```python
#Pie chart of world population percentage by continent - Khadijat

df.groupby('Continent')['World Population Percentage'].sum().plot(kind='pie', autopct='%1.5f%%', figsize=(10, 8))
```
